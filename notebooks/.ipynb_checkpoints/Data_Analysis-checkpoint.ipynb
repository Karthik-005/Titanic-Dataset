```python
df = pd.read_csv("../data/train.csv")
df.sample(5, random_state=5)
```

### Univariate Analysis

#### 1. Categorical Data

#### a. Countpllots

```python
categorical_cols = ['Survived', 'Pclass', 'Sex', 'Embarked']

fig, axes = plt.subplots(2, 2, figsize=(8,6))
fig.tight_layout(pad=0.9)
axes = axes.flatten()

for i, ax in enumerate(axes) :
    sns.countplot(x=df[categorical_cols[i]], ax=ax)

plt.show()
```

#### b. Pie Charts

```python
fig, axes = plt.subplots(2, 2, figsize=(8,6))
fig.tight_layout(pad=0.9)
axes = axes.flatten()

for i, ax in enumerate(axes) :
    df[categorical_cols[i]].value_counts().plot(kind='pie', autopct='%.2f', ax=ax)
    ax.set_title(categorical_cols[i])
plt.show()
```

## 2. Numerical Data

### a. Histograms

```python
num_cols = ["Age", "Fare"]

fig, axes = plt.subplots(1, 2, figsize=(8,6))
fig.tight_layout(pad=2)
axes = axes.flatten()

for i, ax in enumerate(axes) :
    sns.histplot(df[num_cols[i]], bins=15, kde=True, ax= ax) 
    ax.set_title(num_cols[i])
plt.show()    
```

### c. Boxplot

```python
fig, axes = plt.subplots(1, 2, figsize=(8,6))
fig.tight_layout(pad=2)
axes = axes.flatten()

for i, ax in enumerate(axes) :
    sns.boxplot(x=df[num_cols[i]], ax= ax) 
    ax.set_title(num_cols[i])
plt.show()    
```

# Multivariate Analysis

```python
titanic = df
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
iris = sns.load_dataset('iris')
```

```python
tips.sample(5)
```

```python
flights.sample(5)
```

### a. ScatterPlot (Numerical-Numerical)

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='smoker', size='size')
plt.show()
```

### b. BarPlot (Numerical - Categorical)

```python
sns.barplot(x=titanic['Embarked'], y=titanic['Survived'], hue=titanic['Sex'])
plt.show()
```

### c. Boxplot (Numerical-Categorical)

```python
sns.boxplot(x=titanic['Sex'], y=titanic['Age'], hue=titanic['Survived'])
```

### d. DistPlot (Numerical- Categorical)

```python
sns.kdeplot(titanic[titanic['Survived']==1]['Age'], label='Survived')
sns.kdeplot(titanic[titanic['Survived']==0]['Age'], label='Died')
plt.legend()
plt.show()
```

### e. HeatMap (Categorical - Categorical)

```python
sns.heatmap(pd.crosstab(titanic["Pclass"], titanic['Survived']), annot=True)
```

```python
titanic.groupby('Pclass').agg({'Survived':"mean"})
```
