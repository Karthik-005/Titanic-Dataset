# Univariate Analysis

## 1. Categorical Data

### a. Countplot

```python
cat_cols = ["Survived", "Pclass", "Sex", "Embarked"]

fig, axes = plt.subplots(2, 2, figsize=(8,6))
fig.tight_layout(pad=2)
axes = axes.flatten()

for i, ax in enumerate(axes) :
    sns.countplot(data=df, x=cat_cols[i], ax=ax)
plt.show()    
```

### b. Pie Charts

```python
fig, axes = plt.subplots(2, 2, figsize=(8,6))
fig.tight_layout(pad=2)
axes = axes.flatten()

for i, ax in enumerate(axes) :
    df[cat_cols[i]].value_counts().plot(kind='pie', autopct='%.3f', ax=ax)   
    ax.set_title(cat_cols[i])
plt.show()    
```

## 2. Numerical Data

### a. Histograms

```python
num_cols = ["Age", "Fare"]

fig, axes = plt.subplots(1, 2, figsize=(8,6))
fig.tight_layout(pad=2)
axes = axes.flatten()

for i, ax in enumerate(axes) :
    sns.histplot(df[num_cols[i]], bins=15, kde=True, ax= ax) 
    ax.set_title(num_cols[i])
plt.show()    
```

### b. Distplot

### c. Boxplot

```python
fig, axes = plt.subplots(1, 2, figsize=(8,6))
fig.tight_layout(pad=2)
axes = axes.flatten()

for i, ax in enumerate(axes) :
    sns.boxplot(df[num_cols[i]], ax= ax) 
    ax.set_title(num_cols[i])
plt.show()    
```
